```python
#Q1
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Data Preparation
items_csv = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/Items.csv")

customers_csv = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/customers.csv")

order_item_csv = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/order_item.csv")

orders_csv = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/orders.csv")
```

```python
# Explore the Data
tables = {"Items": items_csv, "Customers": customers_csv, "Order_Item": order_item_csv, "Orders": orders_csv}

for name, df in tables.items():
    print(f"\n{name} - info:")
    print(df.info())
```

```python
for name, df in tables.items():
    print(f"\n{name} - describe:")
    print(df.describe(include='all'))
```

```python
for name, df in tables.items():
    print(f"\n{name} - head:")
    print(df.head())
```

```python
# Data Cleaning
items_csv = items_csv.loc[items_csv.notna().sum(axis=1).groupby(items_csv['item_name']).idxmax()]
display(items_csv)
```

```python
# Methods of Operation
items_csv["non_null_count"] = items_csv.notna().sum(axis=1)

items_csv = items_csv.sort_values(by=["item_name", "non_null_count"], ascending=[True, False])

items_csv = items_csv.drop_duplicates(subset="item_name", keep="first")

display(items_csv)
```

```python
# Updating the order_item table
id_map = items_csv.set_index("item_name")["id"].to_dict()
order_item_csv["id"] = order_item_csv["id"].replace({i: id_map[n] for i, n in items_csv[["id", "item_name"]].values})

valid_ids = orders_csv["customer_id"].value_counts()[lambda x: x >= 3].index
customers_csv = customers_csv[customers_csv["id"].isin(valid_ids)]

display(order_item_csv)
```

```python
# Q2
# 3
merged = order_item_csv.merge(items_csv, left_on="item_id", right_on="id")
merged["total_price"] = merged["item_price"] * merged["stock_quantity"]

display(merged)
```

```python
# 4
purchases = merged[["order_id", "item_id", "stock_quantity", "item_price", "total_price"]]
print(purchases)
```

```python
# 5
print("Most expensive purchase:", merged["total_price"].max())
print("Cheapest purchase:", merged["total_price"].min())
print("Average purchase total:", merged["total_price"].mean())
```
